```python
import pandas as pd

# # df = pd.read_csv('2019-09-02indeed_ds_postings.csv')
# names = ['UID', 'job_title', 'company_name', 'city', 'state', 'zipcode', 'neighborhood', 'description', 'salary', 'link']
# df = pd.read_csv('2019-09-02_indeed-ds-postings.csv', index_col='UID', skiprows=1, names=names)

names = ['job_title', 'company_name', 'city', 'state', 'zipcode', 'neighborhood', 'description', 'salary', 'link']
df = pd.read_csv('2019-09-02_indeed-ds-postings.csv', index_col=0, skiprows=1, names=names)
print('Before: ', df.shape)
df = df.drop_duplicates()
df.reset_index(drop=True, inplace=True)
print('After: ', df.shape)

df
# df.loc[0,'description']
```

```python
# df['salary_period'] = df['salary'].apply( lambda s : s.endswith('a year') )
# df[df['salary_period'] == True] = 'year'
# df[ lambda s: df.salarys.endswith('a year') ] 
# df['hi'] = df.salary.apply(lambda s: s.endswith('a year'))
df['salary'] = df['salary'].str.replace(',','')
df['salary'] = df['salary'].str.replace('$','')

df.loc[df['salary'].str.endswith('a year'), 'period'] = 'yearly'
df.loc[df['salary'].str.endswith('a month'), 'period'] = 'monthly'
df.loc[df['salary'].str.endswith('a week'), 'period'] = 'weekly'
df.loc[df['salary'].str.endswith('a day'), 'period'] = 'daily'
df.loc[df['salary'].str.endswith('an hour'), 'period'] = 'hourly'
```

```python
# df['salary'].apply(lambda s: s.rstrip('a year'))
df['salary'] = df['salary'].str.rstrip('a year')
df['salary'] = df['salary'].str.rstrip('a month')
df['salary'] = df['salary'].str.rstrip('a week')
df['salary'] = df['salary'].str.rstrip('a day')
df['salary'] = df['salary'].str.rstrip('an hour')
```

```python
df
```

```python
periods_in_workyear = {'yearly':1, 'monthly':12, 'weekly':52, 'daily':261, 'hourly': 2088}

def get_salary_avg(row):
    sal_range = row['salary']
    sal = sal_range.split('-')
    try:
        avg_salary = (float(sal[0]) + float(sal[1]))/2
        period = row['period']
        avg_salary_yearly = avg_salary * periods_in_workyear[period]
        return avg_salary_yearly
    except:
        try:
            return float(sal_range)
        except:
            return sal_range

df['salary'] = df.apply(get_salary_avg, axis=1)


# df['salary'].apply)
df
```
